##lambda함수 정의

*   이름은 없고 몸체만 있는 함수, 동작에 대한 정의만 있음
*   여러개의 매개변수를 가질 수 있지만 반환값은 하나 이하만 가능하다
*   return 키워드 없이 자동으로 return을 해줌
*   람다식 내부에서 변수 생성 안됨
*   람다식을 변수에 저장할 수 있음(변수 이름(매개변수)이런식으로 사용
*   <font color = "green"> 람다 함수의 장점 코드의 간결함 메모리의 절약




```python
lambda 매개변수 : 표현식
```

<font color ="red"> 익명함수라는 이름처럼 lambda함수는 함수의 이름을 지정하지 않는다.

```python
lambda x : x +1 
```

람다함수는 argument로 받는 숫자에 +1을 해주는 함수이다.

-> 정의와 동시에 사용할 수 있다.

####lambda 함수 예시



*   두 수를 더하는 함수이다



```python
def hap(x,y) :
  return x + y

hap(10,20)
```

*   lambda 함수 바꾸기(한줄로 정의 가능)



```python
(lambda x,y: x + y)(10,20)
```



*   lambda함수 다시 사용



```python
(lambda x : x)(5)
```


1.  괄호 안에 함수표현식을 정의하고 다시 괄호를 써서 값을 넣으면 값이 매개변수에 할당이 됨
2.   실행해서 함수를 처리하면 결괏값이 반환되는데 함수 표현식을 변수에 저장하지 않았으므로 한번 실행됨




```python
x = lambda x,y : x +y
x(5,5)
```

함수표현식을 다시 사용하려면 함수표현식을 변수 x에 인자를 넣고 호출하면 결과가 나옴



## lambda함수 구현





1.  람다 표현식을 사용하기 전에 먼저 숫자를 받은 뒤 10을 더해서 반환하는 함수 plus_ten을 만들어본다.

```python
def plus_ten(x):
  return x +10

plus_ten(1)
```

2.  매개변수 x에 10을 더한 값을 반환하는 간단한 함수이며, lambda함수로 바꾼다.



```python
plus_ten = lambda x: x +10
plus_ten(1)
```

람다 함수식을 보며, lambda x: x +10은 매개변수 x를 하나 받고, x에 10을 더해서 반환한다는 뜻이다






 <u> 매개변수,연산자, 값 등을 조합한 식으로 반환값을 만드는 방식</u>
